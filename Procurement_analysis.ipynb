```python
import pandas as pd

df = pd.read_csv("purchase_order_dataset.csv")

df.head()
```

```python
# Total Spend

df["PO_Value"].sum()
```

```python
# Supplier Spend

supplier_spend = (
    df.groupby("Supplier")["PO_Value"]
      .sum()
      .sort_values(ascending=False)
)

supplier_spend
```

```python
# Supplier Performance

supplier_perf = (
    df.groupby("Supplier")
      .agg({
          "Fill_Rate":"mean",
          "OTIF":"mean",
          "Lead_Time_Days":"mean"
      })
)

supplier_perf
```
